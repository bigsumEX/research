```python
import pickle
# 保存したモデルを読み込む
with open('./output/pkl/app.pkl', 'rb') as file:
    loaded_app = pickle.load(file)

print(loaded_app)
```

```python
from function.SEM import run_SEM
from tqdm import tqdm
import pandas as pd

df_stats = pd.DataFrame()

for i in tqdm(range(len(loaded_app.matrix_list))):
    _, stats, _ = run_SEM(loaded_app.df, loaded_app.matrix_list[i],loaded_app.study.get_trials()[i].params["threshold"])
    # print(dir(stats))
    # display(stats)
    df_stats = pd.concat([df_stats, stats])

```
